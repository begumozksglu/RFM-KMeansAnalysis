## RFM Analizi

```python
print(monetary_df.shape,recency_df.shape,freq_df.shape)
```

```python
#monetary,recency,freq dataframelerini concat fonkisyonu ile tek bir yerde birleştirdik.
rfm=pd.concat([monetary_df,recency_df,freq_df], axis=1)
```

```python
rfm.head()
```

```python
rfm["RecencyScore"] = pd.qcut(rfm["Recency"], 5, labels = [5, 4 , 3, 2, 1])
```

```python
rfm.head()
```

```python
rfm["FrequencyScore"]= pd.qcut(rfm["Frequency"].rank(method="first"),5, labels=[1,2,3,4,5])
```

```python
rfm.head()
```

```python
rfm["MonetaryScore"]=pd.qcut(rfm["Monetary"],5,labels=[1,2,3,4,5])
```

```python
rfm.head()
```

```python
(rfm["RecencyScore"].astype(str)+rfm["FrequencyScore"].astype(str)+rfm["MonetaryScore"].astype(str)).head()
```

```python
rfm["RfmScore"]=(rfm["RecencyScore"].astype(str)+rfm["FrequencyScore"].astype(str)+rfm["MonetaryScore"].astype(str))
```

```python
rfm.head()
```

```python
#rfmscore değeri 555 olan değerleri gözlemliyoruz
#bunun bir nedeni yok gelip gelmiyor mu diye kontrol edildi
rfm[rfm["RfmScore"]=="555"].head()
```

```python
rfm.describe().T
```

```python
# Regular Expressions (Düzenli İfadeler) kullanılarak RFM haritası çıkarıldı
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Loose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
rfm["Segment"]= (rfm["RecencyScore"].astype(str)+rfm["FrequencyScore"].astype(str))
```

```python
rfm.head()
```

```python
#rastgele bir değer yazıp kontrol ediyoruz
rfm[rfm["Segment"]=="11"].head()
```

```python
rfm["Segment"] = rfm["Segment"].replace(seg_map,regex=True)
```

```python
rfm.head()
```

```python
rfm[["Segment","Monetary","Recency","Frequency"]].groupby("Segment").agg(["mean","median","count"])
```

```python
rfm.shape[0]
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(clus)
data_scaled2 = pd.DataFrame(x_scaled)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
plt.figure(figsize=(8,6))
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init='k-means++', n_init=10, max_iter = 300)
    kmeans.fit(data_scaled2)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
inertia_list=[]
silhouette_score_list=[]
for i in range(2,10):
    kmeans.fit(data_scaled2)
    silhouette_score_list.append(silhouette_score(data_scaled2,kmeans.labels_))
    print(silhouette_score_list)
```

```python
kmeans = KMeans(n_clusters = 4, init='k-means++', n_init=10, max_iter=300)
kmeans.fit(data_scaled2)
pred= kmeans.predict(data_scaled2)
```
